```python
# url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# iris = np.genfromtxt(url, delimiter=',', dtype='object')
# names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')

import ssl
import pandas as pd
from urllib.request import urlopen
from io import StringIO

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
context = ssl._create_unverified_context()
data = urlopen(url, context=context)
text = data.read().decode('utf-8')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')
iris_df = pd.read_csv(StringIO(text),header=None,names=names)
```

# Q5 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
bins=[-np.inf,3,5,np.inf]
labels=["small","medium","large"]
iris_df['petal_length_bin']=pd.cut(iris_df['petallength'],bins=bins,labels=labels,right=False)
print(iris_df[['petallength','petal_length_bin']].head())
```
